```python
# 실습 1 - 다중회귀 01: 회귀계수·유의성·R² 해석
from pathlib import Path
import pandas as pd
import statsmodels.formula.api as smf

DATA_DIR = Path('/content/drive/MyDrive/Colab Notebooks')
df = pd.read_csv(DATA_DIR / 'retail_sales_reg1.csv')

# 1. 데이터 확인
print('[데이터 상위 5행]')
print(df.head(), '\n')
print('[기본 정보]')
print(df.info(), '\n')

# 2. 회귀모형 적합 (OLS)
model = smf.ols('sales ~ ad_cost + staff + event_cnt', data=df).fit()

# 3. 결과 요약
print('[회귀 요약 결과]')
print(model.summary())

# 4. 핵심 결과 추출
print('\n[회귀계수]')
print(model.params)
print('\n[p-value]')
print(model.pvalues)
print('\n[R-squared]')
print(model.rsquared)

# 5. 광고비 10단위 증가 효과
beta_ad = model.params['ad_cost']
print('\n광고비 10단위 증가 시 예상 매출 변화:', beta_ad * 10)

# 6. 유의한 변수 목록
sig_vars = model.pvalues[model.pvalues < 0.05].index.tolist()
print('\n유의한 변수:', sig_vars)

# 7. 해석
print('\n[해석]')
print(f'회귀식: sales = {model.params["Intercept"]:.4f} + {model.params["ad_cost"]:.4f}(ad_cost) + {model.params["staff"]:.4f}(staff) + {model.params["event_cnt"]:.4f}(event_cnt)')
if all(model.pvalues[1:] < 0.05):
    print('세 변수의 p-value가 모두 0.05보다 작으므로, 세 변수 모두 월매출에 유의한 영향을 미친다.')
print(f'R² = {model.rsquared:.4f} → 이 모형은 월매출 변동의 약 {model.rsquared*100:.2f}%를 설명한다.')
```

```python
# 실습 2 - 다중회귀 02: 예측값·신뢰구간·오차지표
from pathlib import Path
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.metrics import mean_squared_error

DATA_DIR = Path('/content/drive/MyDrive/Colab Notebooks')
df = pd.read_csv(DATA_DIR / 'weather_reg2.csv')

# 1. 데이터 확인
print('[데이터 상위 5행]')
print(df.head(), '\n')
print('[기본 정보]')
print(df.info(), '\n')

# 2. 회귀모형 적합
model = smf.ols('temperature ~ solar + wind + o3', data=df).fit()

# 3. 결과 요약
print('[회귀 요약 결과]')
print(model.summary())

# 4. 핵심 계수와 p-value
print('\n[회귀계수]')
print(model.params)
print('\n[p-value]')
print(model.pvalues)
print('\n[R-squared]')
print(model.rsquared)

# 5. 특정 조건 예측 (solar=100, wind=5, o3=30)
new_df = pd.DataFrame({'solar': [100], 'wind': [5], 'o3': [30]})
pred = model.predict(new_df)
print('\n[예측값]')
print(pred)

# 6. 신뢰구간 / 예측 요약표
pred_frame = model.get_prediction(new_df).summary_frame(alpha=0.05)
print('\n[예측 요약표]')
print(pred_frame)

# 7. 오차지표 계산
y_true = df['temperature']
y_pred = model.predict(df)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
print('\n[MSE / RMSE]')
print('MSE :', round(mse, 4))
print('RMSE:', round(rmse, 4))

# 8. 해석
print('\n[해석]')
print(f'회귀식: temperature = {model.params["Intercept"]:.4f} + {model.params["solar"]:.4f}(solar) + ({model.params["wind"]:.4f})(wind) + {model.params["o3"]:.4f}(o3)')
print(f'R² = {model.rsquared:.4f} → 기온 변동의 약 {model.rsquared*100:.2f}%를 설명한다.')
print(f'solar=100, wind=5, o3=30 조건 예측 기온: {pred.values[0]:.4f}')
print(f'RMSE = {rmse:.4f} → 모델 예측은 실제 기온과 평균적으로 약 {rmse:.2f} 차이 난다.')
```
